```python
#import dependencies
import matplotlib.pyplot as plt
import statistics
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
#x_axis is a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
#y_axis is a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#creating a line chart using the MATLAB method
plt.plot(x_axis, y_axis)
```

```python
#creating a line chart using the object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#creating a line chart using the object-oriented interface method can be written like this as well:
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#creating a line chart using the object-oriented interface method can be written like this as well:
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#plt.show() used to show figure objects
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')
#create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0, 45)
#create a title
plt.title("PyBer Fare by Month")
#add a grid
plt.grid()
#add a legend
plt.legend()
```

```python
#create the plot and add a label for the legend
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker='*', color='green', linewidth=2, label='Boston')
#create labels for the x and y axis
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#set the y limit between 0 and 45
ax.set_ylim(0, 45)
#create a title
ax.set_title("PyBer Fare by Month")
#add a grid
ax.grid()
#add a legend
ax.legend()
```

```python
#creating a bubble chart using the MATLAB method
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#creating a bubble chart using the MATLAB method, implentingthe for and while loops in the .scatter function
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#create the plot
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color='skyblue', edgecolors='black', linewidth=2, label='Boston')
plt.gca().invert_yaxis()
#create labels for the x and y axis
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#set the y limit between 0 and 50
ax.set_xlim(0, 50)
#create a title
ax.set_title("PyBer Fare by Month")
#add a legend
ax.legend()
```

```python
#creating a pie chart using the MATLAB method
plt.pie(y_axis, labels=x_axis)
```

```python
#creating the plot
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
plt.show()
```

```python
#import statistics dependency
#get the standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
#add error bars to a line chart using the MATLAB method
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#add error bars to a line chart using the object-oriented interface method
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#add error bars to a bar chart using the MATLAB method
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#add error bars to a bar chart using the object-oriented interface method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#changing the major ticks with the MATLAB method
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
#changing the major ticks with the object-oriented interface method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
#adding minor ticks
ax.xaxis.set_minor_locator(MultipleLocator(1))
```
